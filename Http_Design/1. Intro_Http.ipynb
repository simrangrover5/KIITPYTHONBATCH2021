### HTTP

    HTTP --> Hyper Text Transfer Protocol
    
        REQ 
            Head --> Methods
            
            Methods 
                1. GET (default)
                2. HEAD (Head only)
                3. OPTIONS (Allowed Methods)

                2. POST (data --> server side post)
                3. PUT (data --> update)
                4. PATCH (data --> partial update)
                5. DELETE (data --> delete data from server side)
                
         RESPONSE
         
             Head --> Response/Status code
                
             200 --> Successfully done
             404 --> Page not Found
             201 --> Successfully created
