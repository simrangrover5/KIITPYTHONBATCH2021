#### Patterns

```python
n = int(input("\n Enter any number : "))
row = 1
while row<=n:
    col = 1
    while col<=row:
        print("*", end=" ")
        col += 1
    print()
    row += 1
# n = 5, row=1, 1<=5, col=1, 1<=1 --> *, 2<=1 (False)
# row = 2, 2<=5, col=1, 1<=2 --> *, 2<=2 --> *, 3<=2(False)
```

```python
n = int(input("\n Enter any number : "))
row = n
while row>0:
    col = 1
    while col<=row:
        print("*", end=" ")
        col += 1
    print()
    row -= 1
```

```python
n = int(input("\n Enter any number : "))
i = n
while i>0:
    j = n-1
    while j<=i:
        print("*", end=" ")
        j -=1
    print()
    i -= 2
    
# n = 7
```

```python
n = int(input("\n Enter any number : "))
i = n
while i>0:
    j = n-1
    while j<=0:
        print("*", end=" ")
        j -=1
    print()
    i -= 2
    
# n = 7
```

```python
n = int(input("\n Enter any number : "))
i = n
while i>0:
    j = n-i
    while j>0:
        print("*", end=" ")
        j -=1
    print()
    i -= 1
    
# n = 7
# i = 7, 7>0, j = 7-7 = 0, 0>0 (False)
# i = 6, 6>0, j = 7-6 = 1, 1>0(True) --> *
# i = 5, 5>0, j = 7-5 = 2, 2>0 --> *, 1>0 --> *, 0>0
```

```python
n = int(input("\n Enter any number : "))
row = 0
while row<n:
    sp = 0
    while sp<row:
        print(" ", end=" ")
        sp += 1
    star = 0
    while star<(n-row):
        print("*", end=" ")
        star += 1
    print()
    row += 1
```

```python
n = int(input("\n Enter any number : "))
row = 0
while row<n:
    print("  "*row, "* "*(n-row))
    row += 1
```

#### Stone Paper Scissor

      Player1               Player2             Winner
        stone                 scissor             P1
        scissor               paper               P1
        paper                 stone               P1
        
        
     Player2               Player1             Winner
        stone                 scissor             P2
        scissor               paper               P2
        paper                 scissor             P2

```python
print("STONE PAPER SCISSOR GAME".center(80, "#"))
choices = ['stone', 'paper', 'scissor']
print("\n CHOOSE ANY OPTION FROM THESE --> ", *choices)
p1 = input("\n Player 1 --> ").strip().lower()
p2 = input("\n Player 2 --> ").strip().lower()
p1_win = [('stone', 'scissor'), ('scissor', 'paper'), ('paper', 'stone')]
if p1 in choices and p2 in choices:
    if (p1, p2) in p1_win:
        print("\n Player 1 won the game...")
    elif p1 == p2:
        print("\n Game Tie....")
    else:
        print("\n Player 2 won the game...")
else:
    print("\n Player choice is not right...")
print("#"*80)
```
