### Types of Languages 


    1. DDL
        Data Definition Language 
            1. create
                Used to create table or schema
                    SCHEMA --> Over structure of your table 
            2. drop
                Used to drop table and data 
            3. truncate
                Used to delete whole data but not schema
            4. alter
                Change table schema 
            5. rename
                Change table name
            6. comment
                
    2. DML
        Data Manipulation Language 
            1. insert
                To insert data
            2. Delete
                To delete particular row 
            3. Update 
                To change data 
            4. Select
                To select data
    3. DCL
        Data Control Language 
            1. Grant
                Give permission
            2. Revoke
                Take back permission
    4. TCL
        Transactional Control Language 
            1. Commit
                Saves data permanently
            2. Rollback
                It is used to get back the state to a particular point known as 
                savepoint
            3. SavePoint
                Point which will be used to recover data
            4. Set Transaction
                specify characteristics for the transaction
