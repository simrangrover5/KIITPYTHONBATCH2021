```python
def validate_password(passwd):
    if len(passwd) >= 8:
        n, a, s = 0, 0, 0
        for i in passwd:
            if i.isalpha():
                a += 1
            if i.isnumeric():
                n += 1
            if i in " @#$%&^*!.":
                s += 1
        if n>=1 and s>=1 and a>=1:
            return True
        return False
    return False
```

```python
passwd = input("\n Enter password --> ")
validate_password(passwd)
```

```python
def validate_password(passwd):
    if len(passwd) >= 8:
        n, l, u, s = 0, 0, 0, 0
        for i in passwd:
            if i.isupper():
                u += 1
            if i.islower():
                l += 1
            if i.isnumeric():
                n += 1
            if i in " @#$%&^*!.":
                s += 1
        if n>=1 and s>=1 and u>=1 and l>=1:
            return True
        return False
    return False
```

```python
passwd = input("\n Enter password --> ")
validate_password(passwd)
```

    1. We need to have a condition to come out of recursion

```python
def func(n):
    if n<=1:
        return 1
    return n + func(n-1)
```

```python
func(6)
```

```python
def fact(n):
    if n<=1:
        return 1
    return n*fact(n-1)

fact(5)
# fact(5) --> n = 5 --> 5 * fact(4) --> 5 * 24 = 120
# fact(4) --> n = 4 --> 4 *  fact(3) --> 4 * 6 = 24
# fact(3) --> n = 3 --> 3 * fact(2) --> 3 * 2 = 6
# fact(2) --> n = 2 --> 2 * fact(1) --> 2 * 1 = 2
# fact(1) --> n = 1 --> 1
```

```python
fact(7)
```
